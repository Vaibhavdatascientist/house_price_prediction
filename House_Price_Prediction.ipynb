```python
# nv = null values
nv = df.isnull().sum()
nv = nv[nv>0]
nv
```

```python
def null_value_per(data):
    nv = data.isnull().sum()
    nv = nv[nv>0]
    nv_df = pd.DataFrame({'Feature':nv.index,'NullCount':nv.values,
                         'PerNull':(nv.values/data.shape[0])*100})
    nv_df = nv_df.sort_values('PerNull',ascending=False)
    return nv_df
```

#### Feature Selection using T_test (Statistically)

```python
df['Alley'].value_counts()
```

```python
from scipy.stats import ttest_ind   # 2 Sample t-test
```

```python
a = df['Property_Sale_Price'][df['Alley']=='Grvl']
b = df['Property_Sale_Price'][df['Alley']=='Pave']
```

```python
# Ho : There is no statistical significant difference between mean of a and mean of b => Feature is not significant
# H1 : There is statistical significant difference between mean of a and mean of b = > Feature is significant
t_stat,p = ttest_ind(a,b)
print('test_stat',t_stat)
print('p_val',p)
if p>0.05:
    print('Ho is accpted')
else:
    print('H0 is rejected')
```

```python
print(df[df['LotArea']<1881.18].shape)
```

```python
def outlier_treatment_le(x):      # le = lower end
    x = x.clip(lower=x.quantile(0.01))
    return x

def outlier_treatment_ue(x):       # ue = upper end
    x = x.clip(upper=x.quantile(0.99))
    return x
```

```python
df[lower_end] = df[lower_end].apply(outlier_treatment_le)
df[upper_end] = df[upper_end].apply(outlier_treatment_ue)
```

#### Split the data into train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
print(x.shape)
print(2059*0.30)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30, random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train.head()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
def eval_model(ytest,ypred):
    mae = mean_absolute_error(ytest,ypred)
    mse = mean_squared_error(ytest,ypred)
    rmse = np.sqrt(mse)
    r2s = r2_score(ytest,ypred)
    print('MAE',mae)
    print('MSE',mse)
    print('RMSE',rmse)
    print('R2 Score',r2s)
    
def mscore(model):
    print('Train Score',model.score(x_train,y_train))  # Train R2 Score
    print('Test Score',model.score(x_test,y_test))     # Test R2 Score
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr1 = LinearRegression()
lr1.fit(x_train,y_train)  
```

```python
mscore(lr1)

# This is overfit => If Train Score > Test Score


# Underfit => 
# 1) Test Score > Train Score
# 2) Low Variance and High Bias

# Overfit => 
# 1) Test Score < Train Score 
# 2) High Variance and Low Bias

# Ridge and Lasso used to prevent overfitting => we add bias
```

```python
ypred_lr1 =  lr1.predict(x_test)
eval_model(y_test,ypred_lr1)
```

```python
actual_ypred_lr1 = np.exp(ypred_lr1)    

res_df = pd.DataFrame({'Actual_y_test':np.exp(y_test),'Pred':actual_ypred_lr1})
res_df.sample(20)
```
